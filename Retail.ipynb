```python
plt.figure(figsize=(10,6))
sns.countplot(x='Sub-Category',data=df,palette='rocket_r',order=df['Sub-Category'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

**The observations from the above graph**

* The Binders, Paper, furnishings, Phones, storage, art, accessories (maximum)

```python
df['State'].value_counts()
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x='State',data=df,palette='rocket_r',order=df['State'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

**The observations from the above graph are:**

*   The states like **California, New York and Washington** makes the highest nomber of buyers.


```python
#checking for outliers
fig, ax= plt.subplots(1, 2, figsize=(12,8))
sns.boxplot(x=df["Sales"], ax=ax[0])
sns.boxplot(x=df["Profit"], ax=ax[0])
```

```python
#checking for outliers
fig, ax= plt.subplots(1, 2, figsize=(12,8))
sns.boxplot(x=df["Discount"], ax=ax[0])
sns.boxplot(x=df["Quantity"], ax=ax[1])
```

in this case **Sales and Profit** column don't have outliers rather they have a wide range of data. 

But on the other hand **Quantity and Discount** have outliers as seen from the above plot. 

```python
df = df.drop(['Postal Code'],axis = 1)    #dropping postal code columns
```

```python
df.hist(figsize=(10,10),bins=50)
plt.show()
```

* Most customers tends to buy quantity of 2 and 3

* Discount give maximum is 0 to 20 percent

```python
#plotting the Sales and Profit by Category
fig, ax = plt.subplots(1,2, figsize=(12,6))
fig.suptitle("Sales and Profit By Category", fontsize=20)

df_by_category = df.groupby("Category")[["Sales", "Profit", "Region"]].sum()

sns.barplot(x=df_by_category.index, y=df_by_category["Sales" ], ax=ax[0])
sns.barplot(x=df_by_category.index, y=df_by_category["Profit"], ax=ax[1])

fig.tight_layout()
fig.subplots_adjust(top=0.85)
```

 the plot it is observed that while having a reasonable Sales margin the Furniture is gaining a minimun  profit.

```python
#plotting the Sales and Profit by Region
df_groupby_region = df.groupby("Region", sort=True)[["Sales", "Profit"]].sum()

fig, ax = plt.subplots(1 ,2, figsize=(10,5))

ax[0].set_title("Sales by Region")
sns.barplot(x=df_groupby_region.index, y=df_groupby_region["Sales"], ax=ax[0])

ax[1].set_title("Profit by Region")
sns.barplot(x=df_groupby_region.index, y=df_groupby_region["Profit"], ax=ax[1])

plt.tight_layout()
```

* The Western region shows maximum Sales and Profit 

* The South shows the minimun profit

* The Central showing the Lowest Profit among all 

```python
#Count the distribution of Products Catogorically over different Region
plt.figure(figsize=(10, 6))
sns.countplot(x="Region", hue="Category", data=df)
```

the Office Supplies Product shows the Maximum Count.

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['pink','blue'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

* Profit and sales are maximum in consumer segment 
* minimum in Home Office segment

```python
ps = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['blue','orange'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

* highest profit is in california and new york
* least profit is in texas, pennsylvania, Ohio
